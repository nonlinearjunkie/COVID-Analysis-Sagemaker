```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

```python
covid_df=pd.read_csv("s3://covid-data-assignment-w06/owid-covid-data.csv")
print(covid_df.shape)
```

```python
covid_df.head()
```

```python
covid_df.columns
```

```python
imp_columns=['total_cases','new_cases','total_deaths','new_deaths',
        'new_tests', 'total_tests','new_tests_smoothed',
       'tests_per_case', 'positive_rate', 'stringency_index',
       'population', 'population_density', 'median_age', 'aged_65_older',
       'aged_70_older', 'gdp_per_capita', 'extreme_poverty',
       'cardiovasc_death_rate', 'diabetes_prevalence', 'female_smokers',
       'male_smokers', 'handwashing_facilities', 'hospital_beds_per_thousand',
       'life_expectancy', 'human_development_index']
```

```python
covid_df=covid_df[imp_columns]
covid_df.head()
```

```python
covid_df.shape
```

```python
covid_df=covid_df.dropna()
covid_df.shape
```

```python
covid_df.head()
```

```python
Y = covid_df['new_cases']
X=covid_df.drop('new_cases', axis=1)
```

```python
print(X.shape)
```

```python
print(Y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=42)
```

```python
X_train = preprocessing.scale(X_train)
```

```python
model= linear_model.Lasso(max_iter = 10000,alpha=0.1)
model.fit(X_train,y_train)
```

```python
X_test = preprocessing.scale(X_test)
y_predict=model.predict(X_test)
```

```python
print(mean_squared_error(y_test, y_predict))
```
